```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Glucose'], df['Insulin'], alpha=0.6, c=df['Outcome'], cmap='coolwarm', edgecolor='k')
plt.title('Scatter Plot: Glucose vs Insulin', fontsize=14)
plt.xlabel('Glucose')
plt.ylabel('Insulin')
plt.colorbar(label='Outcome (0: No Diabetes, 1: Diabetes)')
plt.grid()
plt.show()
```

```python
#Selecting feature variable which includes all features
X = df.drop(columns=['BloodPressure'])
#Target variables
Y = df['BloodPressure']
```

```python
X.columns
print(Y)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#Scaling all features(continuous + encoded categorical variables b)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Train-test split(80% training, 20% testing)
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size=0.2, random_state=42)
```

```python
#Initializing and training the model
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
#making prediction
Y_pred = model.predict(X_test)
```

```python
#evaluating the model
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```
